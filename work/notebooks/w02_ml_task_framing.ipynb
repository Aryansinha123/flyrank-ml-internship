## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My lane is a scoring (ranking) machine learning problem. The objective is to score webpages based on how likely they need a content refresh. Pages with higher scores should be reviewed first by the SEO team.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target is whether a webpage needs a content refresh. Since there is no direct label, I will use a proxy such as declining traffic, low CTR, or an old publish date to estimate the need for an update.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Since this is a scoring task, Mean Absolute Error (MAE) or RMSE can be used if predicting a refresh score. If the goal is to rank pages, NDCG is also appropriate because it measures how well the most important pages appear at the top of the ranking.A higher NDCG score means the pages that need a content refresh are ranked closer to the top.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
df = pd.read_csv(r"\data\raw\content_refresh_anonymized.csv")
df.head()
df.info()
df.shape
```

The Unit of Analysis

Each row in this dataframe represents one content page (webpage/article). The features describe its SEO performance, engagement, freshness, and traffic, which can be used to estimate whether the page should be refreshed.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

The needs_refresh column is a proxy target created using simple business rules. In a real machine learning project, this target would ideally come from historical data, such as pages that were refreshed and later showed improved performance.

```python

df["needs_refresh"] = (
    (df["days_since_last_update"] > 180) &
    (df["ctr"] < 0.10)
)

df[["days_since_last_update", "ctr", "needs_refresh"]].head()

```

A fixed rule such as "refresh every page older than one year" ignores many important factors like traffic trends, click-through rate, impressions, and content quality. Machine learning can learn how these factors interact and produce a more accurate priority score, allowing SEO teams to focus on the pages that are most likely to benefit from updates.
